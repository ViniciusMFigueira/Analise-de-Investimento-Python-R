```python
## Importando os dados históricos do Indice ISE no Investpy
```

```python
ISE = inv.get_index_historical_data('Corporate Sustainability', country = 'brazil', from_date='01/01/2005', to_date='07/06/2021')
ISE
```

```python
## Criando um Data Frame com os dados dos indices Ibovespa e ISE, utilizando a normalização de dados.
```

```python
Indices_Ibov_x_Ise = pd.DataFrame()
data_inicio = ISE.index[0]

Indices_Ibov_x_Ise['ISE'] = ISE['Close'] / ISE['Close'].loc[data_inicio]
Indices_Ibov_x_Ise['Ibov'] = Ibov['Close'] / Ibov['Close'].loc[data_inicio]
```

```python
## Plotando os gráficos com os preços de fechamento do Índice Bovespa e Ise para comparação de performance

```

```python
Indices_Ibov_x_Ise.plot(figsize=(10,6))
```
